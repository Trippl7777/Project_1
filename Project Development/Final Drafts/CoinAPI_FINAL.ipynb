```python
#Import
import pandas as pd
import numpy as np
import requests as rq
import json
import time

symbols = ["BTC", "ETH", "LTC"] 

#,"XRP","BCHABC", "BNB", "EOS", "XLM", "TRX", "ALGO"]

# Note, only these cryptocurrency datas were determined from multiple pulls to have complete and reliable datasets
# The others, while receiving data, that data was either missing unreasonable amounts of data or could not be found
# ALGO Later traded for LINK as LINK's data was readily available and had higher market volume.
```
